## Matrix Notations, where $\mathbf{A}$ is a matrix $\in \mathbb{R}$ </td></tr>
<table width=100%><col width=20%><col width=80%>
<tr><td style="text-align:left">$\mathbf{A}_{NM}$ </td><td style="text-align:left"> Sparse matrix $\mathbf{A}$ with $N$ rows and $M$ columns </td></tr>
</table>
<table width=100%><col width=20%><col width=80%>
<caption>Operations</caption>
<tr><td style="text-align:left">$\mathbf{A}^T$ </td><td style="text-align:left"> Transpose of matrix $\mathbf{A}$</td></tr>

## Variable Names
<table width=100%><col width=20%><col width=80%>
<caption>General Variables</caption>
<tr> <td style="text-align:left"> $\mathbf{I_N}$ </td>
	<td style="text-align:left"> $N \times N$ identity matrix</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{1_N}$ </td>
	<td style="text-align:left">  $N \times N$ matrix where each entry is $1/N$. </td> </tr>
<tr> <td style="text-align:left"> N </td>
	<td style="text-align:left"> Size of the Data Set, typically $N = n_{samples}$</td> </tr>
<tr> <td style="text-align:left"> M </td>
	<td style="text-align:left"> Size of the Active Set, typically $M\leq N$</td> </tr>
</table>

<table width=100%><col width=20%><col width=80%>
<caption>Data Variables</caption>
<tr> <td style="text-align:left"> $\mathbf{X}$ </td>
	<td style="text-align:left"> Input dataset of size $n_{samples} \times n_{features}$ </td> </tr>
<tr> <td style="text-align:left"> $\mathbf{Y}$ </td>
	<td style="text-align:left"> Properties of the input data, matrix of size $n_{samples} \times n_{properties}$ </td> </tr>
</table>

<table width=100%><col width=20%><col width=80%>
<caption>Eigendecompositions</caption>
<tr> <td style="text-align:left"> $\mathbf{U_A}$ </td>
	<td style="text-align:left"> Set of eigenvectors of a matrix $\mathbf{A}$ represented as columns. <br> Sorted according to their corresponding eigenvalues in decreasing order. </td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\Lambda_A}$ </td>
	<td style="text-align:left"> Diagonal matrix of eigenvalues of matrix $\mathbf{A}$. <br> Sorted in decreasing order from top left to bottom right. <br> Corresponding matrix denoted in a subscript.</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\hat{U}}$ </td>
	<td style="text-align:left"> A truncated set of eigenvectors, typically those corresponding to the largest $n_{PCA}$ eigenvalues in PCA.</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\hat{\Lambda}}$ </td>
	<td style="text-align:left"> Diagonal matrix of the largest eigenvalues, typically truncated to the $n_{PCA}$ largest in PCA.</td> </tr>
</table>

<table width=100%><col width=20%><col width=80%>
<caption>Kernels and Covariance</caption>
<tr> <td style="text-align:left"> $k(x,x')$ </td>
	<td style="text-align:left"> Kernel of $x$ and $x'$</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{K}$ </td>
	<td style="text-align:left"> Kernel matrix between all data points</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{C}$ </td>
	<td style="text-align:left"> Covariance matrix</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\Phi}$ </td>
	<td style="text-align:left"> Input dataset in feature space</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\tilde{\Phi}}$ </td>
	<td style="text-align:left"> Centered input dataset in feature space</td> </tr>
</table>

<table width=100%><col width=20%><col width=80%>
<caption>Projections and Regressions</caption>
<tr> <td style="text-align:left"> $\ell$ </td>
	<td style="text-align:left"> Loss measure</td> </tr>
<tr> <td style="text-align:left"> $\rho$ </td>
	<td style="text-align:left"> Similarity measure</td> </tr>
<tr> <td style="text-align:left"> $\lambda$ </td>
	<td style="text-align:left"> Regularization parameter. </tr>
<tr> <td style="text-align:left"> $\mathbf{T}$ </td>
	<td style="text-align:left"> Latent-space projection</td> </tr>
<tr> <td style="text-align:left"> $\tilde{\mathbf{T}}$ </td>
	<td style="text-align:left"> Whitened latent-space projection such that $\tilde{\mathbf{T}}\tilde{\mathbf{T}}^T = \mathbf{I}$</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{P}_{AB}$ </td>
	<td style="text-align:left"> Projectors from the space of $\mathbf{A}$ to the space of $\mathbf{B}$.</td> </tr>
<tr> <td style="text-align:left"> $PC_i$ </td>
	<td style="text-align:left"> $i^{th}$ Principal Component </td> </tr>
<tr> <td style="text-align:left"> $\mathbf{W}$ </td>
	<td style="text-align:left"> Regression weights</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\hat{Y}}$ </td>
	<td style="text-align:left"> Values predicted from regression</td> </tr>
</table>

<table width=100%><col width=20%><col width=80%>
<caption>Principal Covariates Regression</caption>
<tr> <td style="text-align:left"> $\alpha$ </td>
	<td style="text-align:left"> Projection/regression tuning parameter</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\tilde{K}}$ </td>
	<td style="text-align:left"> Structure space "mixture matrix" between projection and regression. <br> Often involves a factor of $\mathbf{AA}^T$, similar to the kernel.</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{\tilde{C}}$ </td>
	<td style="text-align:left"> Feature space "mixture matrix" between projection and regression. <br> Often involves a factor of $\mathbf{A}^T\mathbf{A}$, similar to the covariance.</td> </tr>
</table>


<table width=100%><col width=20%><col width=80%>
<caption>CUR Decomposition</caption>
<tr> <td style="text-align:left"> $(\mathbf{U}_K, \mathbf{\Sigma}, \mathbf{U}_C)$ </td>
	<td style="text-align:left"> <a href="https://en.wikipedia.org/wiki/Singular_value_decomposition"> SVD </a> decomposition of a matrix</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{A}\approx\mathbf{A}_c \mathbf{SA}_r$ </td>
	<td style="text-align:left">Approximation of a matrix using $c$ columns and $r$ rows</td> </tr>
<tr> <td style="text-align:left"> $\pi_{j}$ </td>
	<td style="text-align:left"> Normalized statistical leverage score. <br> Represents the importance of a given column.</td> </tr>
<tr> <td style="text-align:left"> $\mathbf{P}_c$ </td>
	<td style="text-align:left"> Adjustment of CUR approximation for projecting into latent space.</td> </tr>
</table>
